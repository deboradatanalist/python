###Encontrando e removendo os Outliers

####Movies

```python
df_movies_max = df_movies[df_movies['duration'] == df_movies['duration'].max()]  #fazendo um df com os filmes com duration max
df_movies_max
```

```python
df_movies_min = df_movies[df_movies['duration'] == df_movies['duration'].min()] #fazendo um df com os filmes com duration min
df_movies_min
```

```python
Q1 = df_movies['duration'].quantile(0.25)
Q3 = df_movies['duration'].quantile(0.75)
IQR = Q3 - Q1
print(f'Q1: {Q1}\nQ3: {Q3}\nIQR: {IQR}')
```

```python
outlier_iqr = df_movies[(df_movies['duration'] < (Q1 - 1.5 * IQR)) | (df_movies['duration'] > (Q3 + 1.5 * IQR))] #gerou 286 filmes que so outliers q  11,33% dos df_movies
outlier_iqr
```

```python
df_movies_tratado_iqr = df_movies[(df_movies['duration'] >= (Q1 - 1.5 * IQR)) & (df_movies['duration'] <= (Q3 + 1.5 * IQR))] #excluindo os outliers do df_movies
```

```python
df_movies_tratado_iqr.describe()
```
